```python
#Find which genre is most popular in each country
#Ideas create many different slices with different countries and find the most popular genre in each
# What if multiple countries / genres listed

ex1 = df.dropna(subset=["country","listed_in"]).copy()


ex1["country"] = ex1["country"].str.split(", ")
ex1 = ex1.explode("country")
ex1["listed_in"] = ex1["listed_in"].str.split(", ")
ex1.explode("listed_in")
#display(ex1)
genre_counts = df.groupby(['country', 'listed_in']).size().reset_index(name='count')
most_common_genre = genre_counts.sort_values('count', ascending=False).drop_duplicates('country')
display(most_common_genre)
```

```python
#See which genres are associated with which ratings
#print(df.columns)
ex2 = df.dropna(subset = ["listed_in","rating"]).copy()
ex2["listed_in"] = ex2["listed_in"].str.split(", ")
ex2 = ex2.explode("listed_in")
group = ex2.groupby(["listed_in","rating"]).size().reset_index(name="Count")
mostCommonRatings = group.sort_values(["listed_in","Count"], ascending = [True,False])
Top3 = mostCommonRatings.groupby("listed_in").head(3)
display(Top3)
```

```python
# Visualise when new content added throughout the years
# organising the data to be in th eright forat for matplotlib

ex3 = df.dropna(subset=["date_added"]).copy()
ex3["date_added"] = pd.to_datetime(ex3["date_added"], errors="coerce")

ex3["MY"] = ex3["date_added"].dt.strftime("%B %Y")
display(ex3.sample(5))
groups = ex3["MY"].value_counts().reset_index()
groups["MY_dt"] = (pd.to_datetime(groups["MY"], errors = "coerce"))
groups= groups.sort_values("MY_dt")
display(groups)
#
```

```python
# assigning the x and y axis 
xvalues = groups["MY"]
yvalues = groups["count"]
plt.bar(xvalues,yvalues)
plt.xticks(groups["MY"][::6],rotation=45)
plt.show()
```

```python
#Compare the average duration of older vs newer content
# Define old vs new 
#Split into two different groups and find average

ex4 = df
ex4 = ex4[ex4["type"]=="Movie"].copy()
#display(ex4.head(3))
ex4["duration"] = ex4["duration"].str.replace(" min","")
#display(ex4.head(3))
ex4["duration"] = pd.to_numeric(ex4["duration"], errors="coerce")
#display(ex4.head(3))
oldMovies = ex4[ex4["release_year"]<2020]
newMovies = ex4[ex4["release_year"]>=2020]
oMAverage = oldMovies["duration"].mean()
nMAverage = newMovies["duration"].mean()

display(f"Old movies: {oMAverage}, new movies: {nMAverage}")
display(ex4.sample(10))
```
