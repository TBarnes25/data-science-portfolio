```python
import pandas as pd
```

```python
df = pd.read_csv("netflix_titles.csv")
display(df.head(5))
```

```python
columns = df.columns
display(columns)
```

```python
filtered_movies_df = df[df["type"] == "Movie"]
display(filtered_movies_df.head(5))
```

```python
max = filtered_movies_df["release_year"].max()
min = filtered_movies_df["release_year"].min()
mean = filtered_movies_df["release_year"].mean()
print(f"max: {max} min {min} mean {mean}")
```

```python
filtered_movies_df["duration"] = filtered_movies_df["duration"].str.replace("min", "").str.strip()
display(filtered_movies_df.head(5))
```

```python
filtered_movies_df["duration"] = pd.to_numeric(filtered_movies_df["duration"], errors='coerce')
print(filtered_movies_df["duration"].dtype)
```

```python
max = filtered_movies_df["duration"].max()
min = filtered_movies_df["duration"].min()  
mean = filtered_movies_df["duration"].mean()
print(f"max: {max} min {min} mean {mean}")
```

```python
counts = filtered_movies_df["listed_in"].value_counts()
print(counts)
```
